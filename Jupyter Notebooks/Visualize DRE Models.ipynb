The purpose of this Jupyter notebook is to visualize the doubly robust estimator results. The figure generated here correspond to all parts of Figure 10.

First we import the Python libraries that we need.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

Plotting Functions

```python
def plot_DRE(df_dre):
    plt.figure(figsize=(10, 8))
    plt.rc('xtick',labelsize=20)
    plt.rc('ytick',labelsize=20)
    plt.plot(df_dre['threshold'].values*0.621371, df_dre['output'].values, color='black')
    plt.plot(df_dre['threshold'].values*0.621371, df_dre['lower_CI'].values, color='black', linestyle=':')
    plt.plot(df_dre['threshold'].values*0.621371, df_dre['upper_CI'].values, color='black', linestyle=':')
    plt.ylabel('Treatment effect (trips)', fontsize=25, labelpad=10)
    plt.xlabel('Distance (mi)', fontsize=25, labelpad=10)
    plt.tight_layout()
    plt.show()
```

Plot Boston DRE Model (part a).

```python
df_dre_boston = pd.read_csv('../Data/DRE/boston_combo_DRE_result.csv')
plot_DRE(df_dre_boston)
```

Plot Boston DRE Model (part b).

```python
df_dre_philly = pd.read_csv('../Data/DRE/philly_combo_DRE_result.csv')
plot_DRE(df_dre_philly)
```

Plot Boston DRE Model (part c).

```python
df_dre_dc = pd.read_csv('../Data/DRE/dc_rail_DRE_result.csv')
plot_DRE(df_dre_dc)
```
